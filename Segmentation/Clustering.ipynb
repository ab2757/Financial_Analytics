# Clustering
Clustering is the task of dividing the population or data points into a number of groups such that data points in the same groups are more similar to other data points in the same group than those in other groups. 

In simple words, the aim is to segregate groups with similar traits and assign them into clusters.

Let’s understand this with an example. 

Suppose, you are the head of a rental store and wish to understand preferences of your costumers to scale up your business. Is it possible for you to look at details of each costumer and devise a unique business strategy for each one of them? Definitely not. 


But, what you can do is to cluster all of your costumers into say 10 groups based on their purchasing habits and use a separate strategy for costumers in each of these 10 groups. And this is what we call **clustering**.

# Why use Clustering?
Grouping similar entities together help profile the attributes of different groups. 

In other words, this will give us insight into underlying patterns of different groups. 

There are many applications of grouping unlabeled data, for example, you can identify different groups/segments of customers and market each group in a different way to maximize the revenue. 

Another example is grouping documents together which belong to the similar topics etc. 

Clustering is also used to reduces the dimensionality of the data when you are dealing with a copious number of variables.

# How does Clustering algorithms work?
There are many algorithms developed to implement this technique but the most popular and widely used algorithms in machine learning are :
* K-mean Clustering
* Hierarchical Clustering

# K Means Clustering
K means is an iterative clustering algorithm that aims to find local maxima in each iteration. This algorithm works in these 6 steps :

Step 1 : Specify the desired number of clusters K : Let us choose k=2 for these 5 data points in 2-D space.

<img src = "Images/clustering1.png">

Step 2 : Randomly assign each data point to a cluster : Let’s assign three points in cluster 1 shown using red color and two points in cluster 2 shown using grey color.

<img src = "Images/clustering2.png">

Step 3 : Compute cluster centroids : The centroid of data points in the red cluster is shown using red cross and those in grey cluster using grey cross.

<img src = "Images/clustering3.png">

Step 4 : Re-assign each point to the closest cluster centroid : Note that only the data point at the bottom is assigned to the red cluster even though its closer to the centroid of grey cluster. Thus, we assign that data point into grey cluster

<img src = "Images/clustering4.png">

Step 5 : Re-compute cluster centroids : Now, re-computing the centroids for both the clusters.

<img src = "Images/clustering5.png">

Step 6 : Repeat steps 4 and 5 until no improvements are possible : Similarly, we’ll repeat the 4th and 5th steps until we’ll reach global optima. When there will be no further switching of data points between two clusters for two successive repeats. It will mark the termination of the algorithm if not explicitly mentioned.

# Challenges with K-Means Clustering
There are certain challenges with K-means. 

* Always tries to make clusters of the same size.

* We have to decide the number of clusters at the beginning of the algorithm. 

Ideally, we would not know how many clusters should we have, in the beginning of the algorithm and hence it a challenge with K-means.

This is a gap hierarchical clustering bridges with aplomb. It takes away the problem of having to pre-define the number of clusters. 

# Hierarchical Clustering
Unlike K-mean clustering Hierarchical clustering starts by assigning all data points as their own cluster. As the name suggests it builds the hierarchy and in the next step, it combines the two nearest data point and merges it together to one cluster.

The algorithm works in following steps : 
* Assign each data point to its own cluster.
* Find closest pair of cluster using euclidean distance and merge them in to single cluster.
* Calculate distance between two nearest clusters and combine until all items are clustered in to a single cluster.

In this technique, you can decide the optimal number of clusters by noticing which vertical lines can be cut by horizontal line without intersecting a cluster and covers the maximum distance.

## Segmentation vs Clustering

In Predictive Marketing the term ‘clustering’ gets thrown around quite a lot. It's the predictive marketing version of segmenting. Instead of grouping people, clustering simply identifies what people do most of the time. This allows us to predict what customers are likely to do without boxing them into rigid groups.

Segmenting is the process of putting customers into groups based on similarities, and clustering is the process of finding similarities in customers so that they can be grouped, and therefore segmented. They seem quite similar, but they are not quite the same.

# Outliers Treatment

Below, we have created a function to treat the outliers which are either above 99th percentile or below 1st percentile by clipping those values at 99th and 1st percentile respectively.

```python
#function for treating outliers
def outlier_treatment(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x
    
```

### clip_upper()
Pandas dataframe.clip_upper() is used to trim values at specified input threshold. 

We used this function to trim all the values above the threshold of the input value to the specified input value.

### clip_lower()
Pandas dataframe.clip_lower() is used to trim values at specified input threshold. 

We used this function to trim all the values below the threshold of the input value to the specified input value.

```python
print(data.info())
```

```python
data = data.drop('Cust_id',axis=1)
```

```python
#applying out function to continuous data to treat outliers
print("Before Outlier treatment : ")
print(data.describe())
data_ot = data[['Avg. Yearly Balance_k','Avg Yearly ATM Withdrawals','Avg. Foreign Card Payments','Avg. Digital Payments']].apply(outlier_treatment)
data[['Avg. Yearly Balance_k','Avg Yearly ATM Withdrawals','Avg. Foreign Card Payments','Avg. Digital Payments']]= data_ot
print("After Outlier treatment : ")
print(data.describe())
```

### Code Explanation

* First we are summarizing the original data before outlier treatment.
* Next we select only those variables where we want to apply outlier treatment, i.e., **NUMERIC VARIABLES** and we save this outlier treated data as another variable 'data_ot'.
* Then we replace all original numerical variables data with the outliers treated data(data_ot).
* Next we again summarize the data which is after outlier treatment.

# KMeans Clustering using Scikit-Learn
Scikit-learn's cluster module provides a function KMeans() to do the kmeans clustering.

But before using kmeans function, we need to process the data : 
* All variables should be numeric.
* All variables should be standardized.

### Converting categorical data into numeric by Label Encoding
This is the approach to encode categorical values, which allows you to convert each value in a column to a number. Numerical labels are always between 0 and n_categories-1.

You can do label encoding via attributes .cat.codes on your DataFrame's column.

```python
data['Age'] = data['Age'].astype('category')
data['Age'] = data['Age'].cat.codes

data['Gender'] = data['Gender'].astype('category')
data['Gender'] = data['Gender'].cat.codes

data['Employment Category'] = data['Employment Category'].astype('category')
data['Employment Category'] = data['Employment Category'].cat.codes

data['Income'] = data['Income'].astype('category')
data['Income'] = data['Income'].cat.codes

print(data.head())
```

## Code Explanation

* First we take a column and convert it from object dtype to category dtype.
* Then using cat.codes method we convert the column from category type to numeric by assigning numbers starting from 0,1,.. etc to each category.
* We repeat above two steps for all the columns which we want to convert from object to numerical type.

### astype()
Pandas provide astype() method which is used to cast a pandas object to a specified dtype. astype() function also provides the capability to convert any suitable existing column to categorical type.


### cat.codes
Pandas provide cat.codes method which is used to convert a pandas category type column to numeric type column. This is normally required when data needs to be in complete numeric form to be used for model building.

### Scaling
For this, we will use sklearn's preprocessing module and StandardScaler function.

The StandardScaler assumes your data is normally distributed within each feature and will scale them such that the distribution is now centred around 0, with a standard deviation of 1.

The mean and standard deviation are calculated for the feature and then the feature is scaled based on:
(x – mean(x))/stdev(x)

```python
# standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# statistics of scaled data
pd.DataFrame(data_scaled).describe()
```

### Code Explanation
* Step 1 : Instantiate an object of type StandardScaler()
* Step 2 : Use .fit_transform method to scale the data. Keep in mind that the data after this step is an array
* Step 3 : We converted the data(array type) into DataFrame to summarize it.

```python
from sklearn.cluster import KMeans

# defining the kmeans function with initialization as k-means++
kmeans = KMeans(n_clusters=10, init='k-means++')

# fitting the k means algorithm on scaled data
kmeans.fit(data_scaled)
```

### Code Explanation
* Step 1 : Import KMeans function from sklearn's cluster module.
* Step 2 : Instantiate an object of type KMeans with n_clusters argument equal to number of clusters and init argument equal to 'k-means++'. 

init denotes how to initialize the initial weights of the centroids. 'k-means++' selects initial cluster centers for k-mean clustering in a smart way to speed up convergence. Other methods are to use random or array of initial values.

* Step 3 : Use .fit() method on scaled data to train the kmeans model.

Let’s evaluate how well the formed clusters are. To do that, we will calculate the inertia of the clusters. 

Inertia is nothing but the sum of squared distances of samples to their closest cluster center

```python
# inertia on the fitted data
kmeans.inertia_
```

We got an inertia value of almost 34700. Now, let’s see how we can use the elbow curve to determine the optimum number of clusters in Python.

We will first fit multiple k-means models and in each successive model, we will increase the number of clusters. We will store the inertia value of each model and then plot it to visualize the result:

```python
SSE = []
K = range(1,20)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(data_scaled)
    SSE.append(kmeanModel.inertia_)

# Plot the elbow
plt.subplots(figsize=(10,10))
plt.plot(K, SSE, '^-')
plt.xlabel('k')
plt.ylabel('Sum of Squared Errors')
plt.title('The Elbow Method showing optimum number of clusters:')
plt.grid()
plt.show()
```

Can you tell the optimum cluster value from this plot? Looking at the above elbow curve, we can choose any number of clusters between 10 to 15. 

For simplicity and clear understanding, let’s set the number of clusters as 3 and fit the model:

```python
##Fitting kmeans to the dataset with k=3
km3=KMeans(n_clusters=3,init='k-means++', max_iter=300, n_init=10, random_state=0)
y_means = km3.fit_predict(data)
```

```python
# Let's check the centers of each cluster for each variable
cluster_centers = pd.DataFrame(km3.cluster_centers_)
cluster_centers.columns = data.columns
print(cluster_centers)
```

```python
# To find the labels of the data
# labels are the cluster numbers which have been assigned by the kmeans model to the data
labels=km3.labels_
```

```python
# To convert the labels into a list
labels=labels.tolist()
```

```python
# To convert the labels into a Pandas series
labels=pd.Series(data=labels,index=range(len(labels)))
```

```python
# To check how many data points are assigned to each cluster number(label)
print(labels.value_counts())
```

So, there are 3256 data points belonging to cluster 2, then 3197 points in cluster 0, and so on. This is how we can implement K-Means Clustering in Python.
